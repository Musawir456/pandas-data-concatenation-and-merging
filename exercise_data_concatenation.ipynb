```python

df_movies = pd.read_csv('movies.csv')
df_movies
```

```python
df_finencials = pd.read_csv('financials.csv')
df_finencials
```

```python
#read the csv file: 'new_movies.csv'
df_new_movies = pd.read_csv('new_movies.csv')

#show the dataframe
df_new_movies
```

```python
#concat the two dataframes ['df_movies', df_new_movies]
df_movies = pd.concat([df_movies, df_new_movies], ignore_index = True)

#show the last 5 rows
df_movies.tail()
```

```python
#perform the join operation
df_movies = pd.merge(df_movies, df_languages, on = 'language_id', how = 'inner')

#show the top 5 rows
df_movies.head()
```

```python
# Perform the join (merge) operation
df_movies = pd.merge(df_movies, df_finencials, on='movie_id', how='left')

# Show the last 5 rows
df_movies.tail()
```

```python
#save the dataframe to 'final_complete_data.csv'
df_movies.to_csv('final_complete_data.csv', columns = ['movie_id', 'title', 'lang_name', 'budget', 'revenue', 'currency'], 
                 index = False)
```
